```python
window_text_dict = {}
for text in texts_dict:
  value = texts_dict[text]
  text_len = len(value)
  windowed_text = []
  if text_len>0:
    window = 10
    for i in range(0, text_len, window):
        windowed_text.append(value[i: i+window])
    window_text_dict[text] = windowed_text
```

```python
def get_entities(document):
    # in our case, entities are all unique words
    unique_words = []
    for window in document:
      for word in window:
        if word not in unique_words:
            if word != '':
              unique_words.append(word)
    return unique_words
```

```python
def get_weighted_edges(document):
    import itertools
    # in our case, relations are bigrams in sentences
    # weights are number of equal bigrams
    # use a dict to store number of counts
    bigrams = {}
    for i in range(len(document)):
      for pair in itertools.product(document[i], repeat=2):
        w1, w2 = eval(str(pair))
        if w1 != '' and w2 != '':
            # transform to hashable key in dict
            if str(pair) not in bigrams.keys():
                # weight = 1
                bigrams[str(pair)] = 1
            else:
                # already exists, weight + 1
                bigrams[str(pair)] += 1

    # convert to NetworkX standard form each edge connecting nodes u and v = [u, v, weight]
    weighted_edges_format = []
    for pair, weight in bigrams.items():
        # revert back from hashable format
        w1, w2 = eval(pair)
        weighted_edges_format.append([w1, w2, weight])

    return weighted_edges_format
```

```python
def build_weighted_digraph(document,dict):

    # get graph nodes
    nodes = get_entities(document)

    # get weighted edges
    weighted_edges = get_weighted_edges(document)

    # create graph structure with NetworkX
    G = nx.Graph()
    G.add_nodes_from(nodes)
    G.add_weighted_edges_from(weighted_edges)
    nx.set_node_attributes(G,dict,'attributes')
    G.remove_edges_from(nx.selfloop_edges(G))
    return G
```

```python
generator = sg.mapper.PaddedGraphGenerator(graphs)
gc_model = sg.layer.GCNSupervisedGraphClassification(
    [64, 32], ["relu", "relu"], generator, pool_all_layers=True
)
```

```python
inp1, out1 = gc_model.in_out_tensors()
inp2, out2 = gc_model.in_out_tensors()

vec_distance = tf.norm(out1 - out2, axis=1)
```

```python
pair_model = keras.Model(inp1 + inp2, vec_distance)
embedding_model = keras.Model(inp1, out1)
```

```python
def graph_similarity1(g1, g2):
    matching_graph=nx.Graph()
    attr1 = nx.get_node_attributes(g1,'attributes')
    attr2 = nx.get_node_attributes(g2,'attributes')
    for n1,n2 in g2.edges():
        if g1.has_edge(n1, n2):
            weight1 = ((attr1[n1][0]+attr2[n2][0]))*g1[n1][n2]["weight"]
            weight2 = ((attr1[n1][0]+attr2[n2][0]))*g2[n1][n2]["weight"]
            weight = weight1+weight2
            matching_graph.add_weighted_edges_from([(n1,n2,weight)])

    components = nx.connected_components(matching_graph)
    try :
      largest_component = max(components, key=len)
      subgraph = nx.induced_subgraph(matching_graph, largest_component)
      similarity = np.sum(list(nx.get_edge_attributes(matching_graph,'weight').values()))
    except :
      similarity = 0

    return similarity
```

```python
def calc_precision_recall(doc_sims, relevant):
    cnt = 0
    retrieved = 1
    recall = []
    precision = []
    for doc in doc_sims:
        if doc in relevant:
            cnt += 1
            p = cnt / retrieved
            r = cnt / len(relevant)
            precision += [p]
            recall += [r]
        retrieved += 1

    avg_pre = sum(precision)/len(precision)
    avg_rec = sum(recall)/len(recall)

    #f1 = 2*((avg_pre*avg_rec)/(avg_pre+avg_rec))
    return avg_pre, avg_rec
```
